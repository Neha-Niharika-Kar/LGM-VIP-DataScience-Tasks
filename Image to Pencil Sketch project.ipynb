**Image to Pencil Sketch with Python**



```python
import cv2
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading the original image
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Neha.jpg")
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# converting to grayscale image
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# negating or inverting the grayscale image
inverted = 255 - grayscale
cv2_imshow(inverted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# creating the pencil sketch of the image
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(grayscale, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
